```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./images/morph_dot.png')

#노이즈 없애기
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3 , 3))
erosion = cv2.erode(img, kernel)

merged = np.hstack((img, erosion))


plt.imshow(merged)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./images/morph_hole.png')

#노이즈 없애기
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3 , 3))
erosion = cv2.dilate(img, kernel)

merged = np.hstack((img, erosion))


plt.imshow(merged)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('./images/morph_dot.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('./images/morph_hole.png', cv2.IMREAD_GRAYSCALE)


#노이즈 없애기
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5 , 5))

opening = cv2.morphologyEx(img1, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img2, cv2.MORPH_CLOSE, kernel)

merged1 = np.hstack((img1, opening))
merged2 = np.hstack((img2, closing))
merged3 = np.vstack((merged1, merged2))

plt.figure(dpi=100)
plt.imshow(merged3, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./images/morphological.png')


#노이즈 없애기
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3 , 3))
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

merged = np.hstack((img, gradient))


plt.figure(dpi=100)
plt.imshow(merged)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./images/moon_gray.jpg')


#노이즈 없애기
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (9 , 9))

tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)

merged = np.hstack((img, tophat, blackhat))


plt.figure(dpi=100)
plt.imshow(merged)
plt.axis('off')
plt.show()
```

### 이미지 피라미드

```python
import cv2

img = cv2.imread('./images/cat-01.jpg')[...,::-1]
small = cv2.pyrDown(img)
large = cv2.pyrUp(img)

fig, (ax1, ax2, ax3) =plt.subplots(ncols=3, dpi=150)
ax1.imshow(img)
ax2.imshow(small)
ax3.imshow(large)
plt.show()
```

```python
img = cv2.imread('./images/taekwonv1.jpg')[...,::-1]

small = cv2.pyrDown(img)
large = cv2.pyrUp(small)

laplacian = cv2.subtract(img, large)
restored = large + laplacian

merged = np.hstack((img, laplacian, large, restored))

plt.figure(dpi=200)
plt.imshow(merged)
plt.axis('off')
plt.show()
```

### 허프변환

```python
img = cv2.imread('./images/sudoku.jpg')
img2 = img.copy()
print(img.shape)
h, w = img.shape[:2]

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(img_gray, 100, 200)

lines = cv2.HoughLines(edges, 1, np.pi/180, 130)

for line in lines:
    r, theta = line[0]
    tx, ty  = np.cos(theta), np.sin(theta)
    x0, y0 = tx*r, ty*r
    cv2.circle(img2, (abs(int(x0)), abs(int(y0))), 3, (0, 0, 255), -1)
    x1, y1 = int(x0 + w*(-ty)), int(y0 + h*tx)
    x2, y2 = int(x0 - w*(-ty)), int(y0 - h*tx)
    cv2.line(img2, (x1, y1), (x2, y2), (0, 255, 0), 1)

plt.imshow(img2)
plt.show()

```
