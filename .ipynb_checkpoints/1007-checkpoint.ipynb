```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2. imread('./images/cat-01.jpg')
raw = img.copy()
rows, cols = img.shape[:2]
print(rows, cols)

pts1 = np.array([
    [100, 50], [200, 50], [100, 200]    
],dtype=np.float32)

pts2 = np.array([
    [80, 70], [210, 60], [250, 120]    
],dtype=np.float32)

cv2.circle(img, (100, 50), 5, (255,0,0), -1) #-1 다 채워넣겠다
cv2.circle(img, (200, 50), 5, (0 , 255, 0), -1)
cv2.circle(img, (100, 200), 5, (0 , 0, 255), -1)

matrix = cv2.getAffineTransform(pts1, pts2)
dst = cv2.warpAffine(img, matrix, (int(cols*1.5), rows))

fig, (ax1, ax2) = plt.subplots(ncols=2, dpi =150)
ax1.imshow(raw[...,::-1])
ax2.imshow(dst[...,::-1])
fig.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2. imread('./images/cat-01.jpg')
raw = img.copy()
rows, cols = img.shape[:2]
print(rows, cols)

pts1 = np.array([
    [0, 0], [0, rows], [cols, 0]    , [cols, rows]
],dtype=np.float32)

pts2 = np.array([
    [100, 50], [10, rows-50], [cols-100, 50], [cols-50, rows-50]
],dtype=np.float32)

matrix = cv2.getPerspectiveTransform(pts1, pts2)
#print(matrix)
dst = cv2.warpPerspective(img, matrix, (cols, rows)) #내부에 있는 이미지 사용, 1.5배 읺더 더;ㄴ디


matrix = cv2.getPerspectiveTransform(pts1, pts2)

fig, (ax1, ax2) = plt.subplots(ncols=2, dpi =150)
ax1.imshow(raw[...,::-1])
ax2.imshow(dst[...,::-1])
fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io, color

img = io.imread('./images/cat-01.jpg')
img = color.rgb2gray(img)
io.imshow(img)
w, h  = img.shape


mat_identity = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])

reflect = np.array([
    [1, 0, 0],
    [0, 1, -h],
    [0, 0, 1]
])

#matrix곱해주기
result = np.array([
    [1, 0, 0],
    [0, -1, 0],
    [0, 0, 1]
])@reflect

img1 = ndimage.affine_transform(img, result)
plt.imshow(img1, cmap ='gray')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.transform import ProjectiveTransform

src = io.imread('./images/cat-01.jpg')
h, w, c = src.shape
print(h, w, c)

dst = np.zeros((h, w, c))

pt = ProjectiveTransform()
pts1 = np.array([
    [0. , 0.], [h-1, 0], [h-1, w-1],[0,w-1]
], dtype=np.float32)

pts2 = np.array([
    [105. , 74.], [340, 46], [300, 377],[60, 322]
], dtype=np.float32)
print(pt.estimate(pts1, pts2))

x, y =np.mgrid[:h, :w]

dst_indices =np.hstack((x.reshape(-1,1), y.reshape(-1, 1)))
src_indices = np.round(pt.inverse(dst_indices),0).astype(int)
valid_idx =np.where(
    (src_indices[:,0] < h)
    &(src_indices[:,1] < w)
    &(src_indices[:,0] >= 0)
    &(src_indices[:,1] >= 0)
)

dst_indices_valid = dst_indices[valid_idx]
src_indices_valid = src_indices[valid_idx]

dst[dst_indices_valid[:, 0], dst_indices_valid[:, 1]] = \
    src[src_indices_valid[:, 0], src_indices_valid[:, 1]]
dst = dst.astype(np.uint8)

plt.imshow(dst)
io.imsave('./images/transform_cat.jpg',dst)
```
